```python
import pandas as pd
from surprise import SVD, SVDpp, Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
reader = Reader(rating_scale=(ratings['rating'].min(), ratings['rating'].max()))
reader
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
data
```

```python
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
svd = SVD()
svd.fit(trainset)
```

```python
svdpp = SVDpp()
svdpp.fit(trainset)
```

```python
svd_predictions = svd.test(testset)
svdpp_predictions = svdpp.test(testset)
```

```python
svd_mse = accuracy.mse(svd_predictions)
svdpp_mse = accuracy.mse(svdpp_predictions)
```

```python
print(f"SVD MSE: {svd_mse}")
print(f"SVD++ MSE: {svdpp_mse}")
```
